```python
z = df_sp500[df_sp500.tic.isin(['AAPL'])]
z['return'] = z.close.pct_change()
z = z.dropna(how='all',axis=0)
```

```python
pv = z.pivot(values='return',columns='tic',index='date')
```

```python
pct = pv['AAPL']

def sharpe(ls):
    return np.mean(ls)/np.std(ls)

# ls1 contains the true SR values
ls1 = []
sr0 = sharpe(pct[:200])
for i in range(500):
    sr = sharpe(pct[:200+i+1])
    ls1.append(sr-sr0)
    sr0 = sr

# ls2 uses cumulated DSR to approximate SR 
ls2 = []
eta = 0.004
# use the first 200 days to set an initial value of SR
sr = sharpe(pct[:200]) 
for i in range(500):
    A = np.mean(pct[:200+i])
    B = np.mean(pct[:200+i]**2)
    delta_A = pct[200+i+1] - A
    delta_B = pct[200+i+1]**2 - B
    Dt = (B*delta_A - 0.5*A*delta_B) / (B-A**2)**(3/2)
    sr += eta * Dt
    ls2.append(Dt*eta)
```

```python
import numpy as np
pct = pv['AAPL']
def sharpe(ls):
    return np.mean(ls)/np.std(ls)

# ls1 contains the true SR values
ls1 = []
sr0 = sharpe(pct[:200])
for i in range(500):
    sr = sharpe(pct[:200+i+1])
    ls1.append(sr)
    sr0 = sr

# ls2 uses cumulated DSR to approximate SR 
ls2 = []
eta = 0.01
# use the first 200 days to set an initial value of SR
sr = sharpe(pct[:200]) 
for i in range(500):
    A = np.mean(pct[:200+i])
    B = np.mean(pct[:200+i]**2)
    delta_A = pct[200+i+1] - A
    delta_B = pct[200+i+1]**2 - B
    Dt = (B*delta_A - (0.5*A*delta_B)) / (B-A**2)**(3/2)
    sr += eta * Dt
    ls2.append(Dt*eta)
```

```python
import matplotlib.pyplot as plt
import scienceplots
%matplotlib inline
with plt.style.context('science','ieee'):
    plt.figure(figsize=(9, 4),dpi=200) 
    ax = plt.subplot(1, 2, 1)
    plt.plot(ls1)
    plt.title('Sharpe Ratio')
    ax.set_xlabel('time')
    ax.set_ylabel('Sharpe Ratio')
    ax = plt.subplot(1, 2, 2)
    plt.plot(ls2)
    plt.title('Sharpe Ratio')
    ax.set_xlabel('time')
    plt.title('Differential Sharpe Ratio')
    plt.show()
```
