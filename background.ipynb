```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Load the Iris dataset
iris = sns.load_dataset('iris')

# Calculate the covariance matrix
cov_matrix = iris.iloc[:, :-1].cov()

# Calculate the correlation matrix
corr_matrix = iris.iloc[:, :-1].corr()

# Set up the plot
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# Plot the covariance matrix
sns.heatmap(cov_matrix, annot=True, cmap='coolwarm', ax=axes[0], fmt='.2f', cbar=True)
axes[0].set_title('Covariance Matrix')

# Plot the correlation matrix
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', ax=axes[1], fmt='.2f', cbar=True)
axes[1].set_title('Correlation Matrix')

# Adjust layout to prevent clipping of titles/labels
plt.tight_layout()

# Save the figure as an image file (e.g., PNG)
plt.savefig('cov_corr_matrices.png', dpi=300)

# Show the plot (optional)
plt.show()
```
